# The following three functions:
        
### 1: Gets an individual audience rating for a given movie url
### 2: Gets all of the audience ratings (100) for a list of urls in a year
### 3: Gets all of the titles, ranks, ratings, and years for the top 100 movies in the given year

```python
def get_aud_rating(url):
    html_page = requests.get(f'https://www.rottentomatoes.com{url}')
    soup = BeautifulSoup(html_page.content, 'html.parser')
    return int(soup.findAll('span', class_="mop-ratings-wrap__percentage")[1].text.strip()[:-1]) #audience rating
```

```python
def strip_audience_ratings(year):    
    html_page = requests.get(f'https://www.rottentomatoes.com/top/bestofrt/?year={year}')
    soup = BeautifulSoup(html_page.content, 'html.parser')
    table = soup.find('table', class_='table')
    table_rows = table.findAll('tr')
    
    urls = []
    for i in table_rows[1:]:
        urls.append(i.find('a')['href'])

    ratings = []
    for url in urls:
        try:
            ratings.append(get_aud_rating(url))
        except:
            ratings.append('Nan')
    
    return ratings
```

```python
def strip_rotten(year):    
    html_page = requests.get(f'https://www.rottentomatoes.com/top/bestofrt/?year={year}')
    soup = BeautifulSoup(html_page.content, 'html.parser')
    table = soup.find('table', class_='table')
    table_rows = table.findAll('tr')

    titles = []
    for i in table_rows[1:]:
        titles.append(i.find('a').text.strip(f'({year})').strip())

    rank = []
    for i in table_rows[1:]:
        rank.append(int(i.find('td', class_='bold').text[:-1]))

    rating = []
    for i in table_rows[1:]:
        rating.append(int(i.find('span', class_='tMeterScore').text.strip()[:-1]))
        
    years = []
    for i in range(0,100):
        years.append(f'{year}')
        
    aud_rating = strip_audience_ratings(year)
    
    df = pd.DataFrame([rank,titles,rating,years,aud_rating]).transpose()
    df.columns = ['rank', 'title', 'rating','year','audience_rating']
    
    return df
```
