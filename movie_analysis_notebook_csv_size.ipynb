```python
def strip_rotten(year):    
    html_page = requests.get(f'https://www.rottentomatoes.com/top/bestofrt/?year={year}')
    soup = BeautifulSoup(html_page.content, 'html.parser')
    table = soup.find('table', class_='table')
    table_rows = table.findAll('tr')

    titles = []
    for i in table_rows[1:]:
        titles.append(i.find('a').text.strip(f'({year})').strip())

    rank = []
    for i in table_rows[1:]:
        rank.append(int(i.find('td', class_='bold').text[:-1]))

    rating = []
    for i in table_rows[1:]:
        rating.append(int(i.find('span', class_='tMeterScore').text.strip()[:-1]))
        
    years = []
    for i in range(0,100):
        years.append(f'{year}')
    
    df = pd.DataFrame([rank,titles,rating,years]).transpose()
    df.columns = ['rank', 'title', 'rating','year']
    
    return df
```

```python
df_2019 = strip_rotten(2019)
```

```python
df_2018 = strip_rotten(2018)
df_2017 = strip_rotten(2017)
df_2016 = strip_rotten(2016)
df_2015 = strip_rotten(2015)
df_2014 = strip_rotten(2014)
df_2013 = strip_rotten(2013)
df_2012 = strip_rotten(2012)
df_2011 = strip_rotten(2011)
df_2010 = strip_rotten(2010)
```

```python
big_df = pd.concat([df_2010, df_2011, df_2012, df_2013, df_2014, df_2015, df_2016, df_2017, df_2018, df_2019])
```
